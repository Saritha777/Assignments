```python
#replacing missing values with Not a Number(NaN)
df=data.replace(0,np.NaN)
print(df.isnull().sum())
print("\nshape ",df.shape)
print("\n",df.head(10))
```

```python
#filling missing values using mean,forward and backward filling 
mean_value=data['Age'].mean()
data['Age']=data['Age'].fillna(mean_value)
data.fillna(method='bfill')
data.fillna(method='pad')
data.head(10)
```
