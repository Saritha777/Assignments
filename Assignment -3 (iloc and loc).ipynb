```python
import pandas as pd
df=pd.read_csv("Desktop\\python programs\\nba.csv")
print(df)
```

# iloc 

```python
#extracting single row
r1=df.iloc[5]
print(r1)
#extracting multiple rows with index
r2=df.iloc[4:8]
print(r2)
#extracting particular rows and columns
r3=df.iloc[0:3,0:3]
print(r3)
```

# loc

```python
r1=df.loc[1]
print(r1)
#extracting using mutiple parameters
r2=df.loc[[1,2]]
print(r2)
#extracting mutiple rows with same index we can change index column in loc
df=pd.read_csv("Desktop\\python programs\\nba.csv",index_col="Position")
r3=df.loc["PG"]
print(r3)
#extracting rows between two index labels
df=pd.read_csv("Desktop\\python programs\\nba.csv")
r4=df.loc[0:20]
print(r4)
```

```python
#comparing iloc and loc
import pandas as pd
df=pd.read_csv("Desktop\\python programs\\nba.csv")
r1=df.iloc[3]
r2=df.loc[3]
print(r1==r2,"\n\n\n") 
#if data have not a number value then it gives false
r3=df.iloc[5]
r4=df.loc[5]
print(r3==r4)
```
