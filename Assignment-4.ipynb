# Date Formats

```python
from datetime import date
x=date.today()
print(x.strftime("%Y"))# year full version
print(x.strftime("%y"))#year short version
print(x.strftime("%m"))# month number
print(x.strftime("%M"))# Minutes
print(x.strftime("%d"))#day of the month
print(x.strftime("%a"))#Week day short version
print(x.strftime("%A"))#Week day full version
print(x.strftime("%b"))#Month short version
print(x.strftime("%B"))#Month full version
print(x.strftime("%c"))#local date and time
print(x.strftime("%f"))#Micro seconds
print(x.strftime("%H"))#Hours(0-23)
print(x.strftime("%I"))#hour()
print(x.strftime("%j"))#day of the year
print(x.strftime("%p"))#am/pm
print(x.strftime("%S"))#seconds
print(x.strftime("%U"))#week number of the year starting from sunday
print(x.strftime("%w"))#weekday number of the week monday=1
print(x.strftime("%W"))#week number starting from monday
print(x.strftime("%x"))#local date 
print(x.strftime("%X"))#local time
#print(x.strftime("%z"))
#print(x.strftime("%Z"))
```

# join and concat command

```python
import pandas as pd
df1=pd.read_csv("Desktop\\python programs\\Datasets\\nba.csv")
df=df1.sort_values('Age',ascending=True)
h=df.head(10)
t=df.tail(10)
A1=pd.concat([h,t])#concat command
print("Concat\n",A1)
#merge using join
x=df[['Name','Age','College']]
y=df[['Age','Team','Salary','Number']]
h1=x.head(10)
t1=y.tail(10)
A2=pd.merge(h1,t1,how='outer',on='Age')
print("\nOuter join\n",A2)
A3=pd.merge(h1,t1,how='inner',on='Age')
print("\nInner join\n",A3)
A4=pd.merge(h1,t1,how='left',on='Age')
print("\nLeft join\n",A4)
A5=pd.merge(h1,t1,how='right',on='Age')
print("\nRight join\n",A5)
```
